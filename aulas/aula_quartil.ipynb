### MEDIANA - valor central de um conjunto ordenado
### QUARTIS - divide em 4 partes
### Q1 - 25%
### Q2 - 50%
### Q3 - 75% 

### CRTL + D + segurar = selecionar todos de uma vez

```python
import numpy as np
import pandas as pd
```

```python
dados =  np.array([12,15,17,20,22,25,28,30,35,40])
mediana = np.median(dados)
q1 = np.percentile(dados, 25)
q2 = np.percentile(dados, 50)
q3 = np.percentile(dados, 75)
print((f'primeiro quartil: {q1}'))
print((f'mediana: {mediana}'))
print((f'mediana: {q2}'))
print((f'terceiro quartil: {q3}'))
```

```python
tempo_espera_array = np.array(df['Tempo de Espera (min)'])
mediana = np.median(tempo_espera_array)
q1 = np.percentile(tempo_espera_array, 25) ### equivalente ao primeiro quartil, que menos esperaram
q2 = np.percentile(tempo_espera_array, 50)
q3 = np.percentile(tempo_espera_array, 75)
print((f'primeiro quartil: {q1}'))
print((f'mediana: {mediana}'))
print((f'mediana (segundo quartil): {q2}'))
print((f'terceiro quartil: {q3}'))
```
