```python
import matplotlib.pyplot as plt
import numpy as np  
import pandas as pd
```

```python
df = pd.DataFrame({
    'ano': [2010, 2011, 2012, 2013, 2014, 2015],
    'vendas': [100, 120, 300, 260, 400, 450]

})
plt.plot(df['ano'], df['vendas'])
plt.title('vendas')
plt.xlabel('ano') ### não adicionar titulos para os eixos x e y. Má pratica
plt.ylabel('vendas')
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame({
    'Produto': ['A', 'B', 'C', 'D'],
    'Vendas': [300, 450, 150, 500]})

df = df.sort_values(by='Vendas', ascending=False)### recebimento de nova caracteristica - usar o inplace=True para facilitar

plt.bar(df['Produto'], df['Vendas'])
plt.title('Vendas por Produto')
plt.xlabel('Produto')
plt.ylabel('Vendas')
plt.show()
```

```python
dados = {
    'Ano': [2018, 2019, 2020, 2021, 2022],
    'Vendas': [150, 200, 250, 300, 350],

}

df = pd.DataFrame(dados)

plt.figure(figsize=(8, 5))
plt.barh(df['Ano'], df['Vendas'], color='darkred')
plt.title('Vendas por Ano')
plt.xlabel('Vendas')
plt.ylabel('Ano')
plt.savefig('grafico_barras.horizontal.png')
plt.show()
```

```python
df_dias = pd.DataFrame({'dias': ['segunda', 'terca', 'quarta', 'quinta', 'sexta', 'sabado', 'domingo']
      , 'vendas': [150, 200, 250, 300, 350, 400, 450]
})


df_produto = pd.DataFrame({'Produto': ['A', 'B ', 'C', 'D', 'E'],
    'venda': [120, 90 ,150, 300, 500]
})

fig, axs = plt.subplots(2,2, figsize = (12,8))

axs[0,0].plot(df_dias['dias'],df_dias['vendas'], marker = 'o', linestyle = '-', color = 'red')
axs[0,0].set_title('Vendas por dias')
axs[0,0].set_xlabel('dias')
axs[0,0].set_ylabel('vendas')

axs[0,1].bar(df_produto['Produto'],df_produto['venda'], color = 'blue')
axs[0,1].set_title('Vendas por produto')
axs[0,1].set_xlabel('Produto')
axs[0,1].set_ylabel('venda')

axs[1,0].barh(df_produto['Produto'],df_produto['venda'], color = 'blue')
axs[1,0].set_title('Vendas por produto')
axs[1,0].set_xlabel('Produto')
axs[1,0].set_ylabel('venda')


axs[1,1].axis('off')

axs[1,1].text(0.1, 0.8, 'Resumo aula:', fontsize=14, color = 'black')
axs[1,1].text(0.1, 0.6, '- subplotes usa matriz de eixos', fontsize=12, color = 'black')
axs[1,1].text(0.1, 0.4, '- mais controle sobre cada gráfico', fontsize=12, color = 'black')
axs[1,1].text(0.1, 0.2, '- ideal para projetos mais complexos', fontsize=12, color = 'black')

plt.tight_layout()

```
