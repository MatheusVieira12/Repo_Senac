```python
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv('preco_semestral.csv', sep=';')
df["Valor de Venda"] = df["Valor de Venda"].astype(str).str.replace(",", ".")
df
```

```python
df_gasolina = df[(df['Produto'] == 'GASOLINA') & (df['Estado - Sigla'] == 'MG')]
df_gasolina["Valor de Venda"] = pd.to_numeric(df_gasolina["Valor de Venda"], errors="coerce")
```

```python
df_gasolina_ordenado = df_gasolina.sort_values(by="Valor de Venda", ascending=True)
(df_gasolina_ordenado.head(5))

```

```python
df_gasolina_media = df_gasolina.groupby("Estado - Sigla")["Valor de Venda"].mean().reset_index()
df_gasolina_media
```
