```python
import pandas as pd
import numpy as np

Func_dados = pd.read_csv('funcionarios_dados_ti.csv', sep=';')
Func_dados.head()
```

```python
soma_sal = Func_dados.groupby('Cargo')['Salario'].sum().reset_index()
soma_sal
```

```python
func = np.array(Func_dados['Salario'])
mediana_sal = np.median(func)
media_sal = np.mean(func).round(1)  
q_1 = np.percentile(func, 25) ### equivalente ao primeiro quartil, que menos esperaram
q_2 = np.percentile(func, 50)
q_3 = np.percentile(func, 75)
print((f'media de salarios: {media_sal}'))
print((f'primeiro quartil: {q_1}'))
print((f'mediana: {mediana_sal}'))
print((f'mediana (segundo quartil): {q_2}'))
print((f'terceiro quartil: {q_3}'))
distancia = (media_sal - mediana_sal) / mediana_sal
print((f'distancia entre media e mediana: {round(distancia * 100, 2)}'))
```
