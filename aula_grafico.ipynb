```python
import matplotlib.pyplot as plt
import numpy as np  
import pandas as pd
```

```python
df = pd.DataFrame({
    'ano': [2010, 2011, 2012, 2013, 2014, 2015],
    'vendas': [100, 120, 300, 260, 400, 450]

})
plt.plot(df['ano'], df['vendas'])
plt.title('vendas')
plt.xlabel('ano') ### não adicionar titulos para os eixos x e y. Má pratica
plt.ylabel('vendas')
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame({
    'Produto': ['A', 'B', 'C', 'D'],
    'Vendas': [300, 450, 150, 500]})

df = df.sort_values(by='Vendas', ascending=False)### recebimento de nova caracteristica - usar o inplace=True para facilitar

plt.bar(df['Produto'], df['Vendas'])
plt.title('Vendas por Produto')
plt.xlabel('Produto')
plt.ylabel('Vendas')
plt.show()
```

```python
dados = {
    'Ano': [2018, 2019, 2020, 2021, 2022],
    'Vendas': [150, 200, 250, 300, 350],

}

df = pd.DataFrame(dados)

plt.figure(figsize=(8, 5))
plt.barh(df['Ano'], df['Vendas'], color='darkred')
plt.title('Vendas por Ano')
plt.xlabel('Vendas')
plt.ylabel('Ano')
plt.savefig('grafico_barras.horizontal.png')
plt.show()
```

```python
df = {'dias': ['segunda', 'terca', 'quarta', 'quinta', 'sexta', 'sabado', 'domingo']
      , 'vendas': [150, 200, 250, 300, 350, 400, 450]
      }

dias = df['dias']
valores = df['vendas']

plt.plot(dias, valores, marker='o', linestyle='-', color='green')
plt.title('Vendas por Dia da Semana')### grafico de linha ver o dado ao longo do tempo
plt.xlabel('Dias da Semana')
plt.ylabel('Vendas')
plt.show()
```
